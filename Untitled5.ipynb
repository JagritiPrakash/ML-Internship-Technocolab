```python
import pandas as pd
dataset = pd.read_csv("parkinsons.csv")
```

```python
dataset.tail()
```

```python
x1=dataset.drop(['status','name'],axis=1)
y1=dataset['status']
```

```python
from sklearn.preprocessing import MinMaxScaler
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
scaler=MinMaxScaler((-1,1))
x2=scaler.fit_transform(x1)
y2=y1
```

```python
x2_train,x2_test,y2_train,y2_test=train_test_split(x2, y2, test_size=0.2, random_state=5)
```

```python
model=XGBClassifier()
model.fit(x2_train,y2_train)
```

```python
y2_pred=model.predict(x2_test)
print(accuracy_score(y2_test, y2_pred)*100)
```

```python
def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn
warnings.filterwarnings("ignore", category=DeprecationWarning)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler((-1, 1))
x = scaler.fit_transform(x)
y = y
```

```python

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,test_size=0.2,random_state=5)
```

```python
model = XGBClassifier()
model.fit(x_train, y_train)
```

```python
y_hat = [round(yhat) for yhat in model.predict(x_test)]
print(accuracy_score(y_test, y_hat)) # Test set accuracy
y_hat = [round(yhat) for yhat in model.predict(x)]
print(accuracy_score(y, y_hat)) # Full set accuracy
```
